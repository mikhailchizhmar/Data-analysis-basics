```python
import pandas as pd, matplotlib.pyplot as plt
from randan.bivariate_association import Crosstab
from randan.descriptive_statistics import ScaleStatistics
```

```python
df=pd.read_excel('FL_Конкурсы_20231112.xlsx', index_col=0)
df
```

```python
# H0: связь между переменными "Разделы" и "Участники" отсутствует
# H1: связь между переменными "Разделы" и "Участники" есть

# доверительная вероятность 95%
```

```python
for i in df.index:
    try:
        df.loc[i, "Раздел_общий"] = df.loc[i, "Разделы"].split("/")[0]
    except:
        df.loc[i, "Раздел_общий"] = df.loc[i, "Разделы"]
```

```python
ctab = Crosstab(df, "Раздел_общий", "Участники")
```

```python
# 100% ячеек имеют ожидаемые частоты меньше 5. Таблица сопряженности очень разрежена, => полученным выводам доверять нельзя
```

```python
ss = ScaleStatistics(df, ['Участники'])
```

```python
participants_ranges = [round(ss.summary()['25%']['Участники'], 0),
                      round(ss.summary()['median']['Участники'], 0),
                      round(ss.summary()['75%']['Участники'], 0),
                      round(ss.summary()['max']['Участники'], 0)]
participants_ranges
```

```python
participants_labels = ["1) До 8 уч - Наименее популярные",
                      "2) До 17 уч - Слабо популярные",
                      "3) До 44 уч - Популярные",
                      "4) До 141 уч - Наиболее популярные"]
```

```python
df.loc[df['Участники'] <= participants_ranges[0], 'Диапазон_участников'] = participants_labels[0]
```

```python
for i in range(1, len(participants_ranges)):
    df.loc[(df['Участники'] > participants_ranges[i-1]) & (df['Участники'] <= participants_ranges[i]), 
            'Диапазон_участников'] = participants_labels[i]
df
```

```python
ctab = Crosstab(df, "Раздел_общий", "Диапазон_участников")
```

```python
# p-value очень большое (намного больше 0.05), в связи с чем делаем вывод о том, что связи между переменными "Разделы" и "Участники" не наблюдается
# Более того, 
```
