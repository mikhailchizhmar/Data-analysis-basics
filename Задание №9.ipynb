```python
import pandas as pd, matplotlib.pyplot as plt
from randan.bivariate_association import Correlation
```

```python
df=pd.read_excel('FL_Конкурсы_20231112.xlsx', index_col=0)
df
```

```python
# H0: связь между переменными "Число комментариев" и "Кандидаты" отсутствует => коэф. Пирсона = 0
# H1: связь между переменными "Число комментариев" и "Кандидаты" есть => коэф. Пирсона не равен 0
```

```python
corr = Correlation(df[['Число комментариев', 'Кандидаты']]) # коэф корреляции Пирсона
# доверительная вероятность 95%
```

```python
corr.sort_correlations()
# коэффициент корреляции = 0.106, => корреляция слабая прямая
# p-value намного больше 0.05, поэтому не можем доверять этому коэффициенту
```

```python
var = ["Число комментариев", "Кандидаты"]
plt.figure(figsize=(8,6))
plt.scatter(df[var[0]], df[var[1]], c='red') # График для иллюстрации парной связи (сначала ось х, потом ось у)
plt.title(f'Корреляция между переменной {var[0]} и переменной {var[1]}') # Заголовок графика
plt.xlabel(var[0]) # Название оси Х
plt.ylabel(var[1]);
```

```python
# так как большинство точек располагаются параллельно оси х, делаем вывод о почти нулевой корреляции
# линейная связь здесь не ярко выражена, поэтому воспользуемся другим методом
```

```python
# посмотрим на коэффициент корреляции Спирмана
corr2 = Correlation(df[['Число комментариев', 'Кандидаты']], method='spearman')
```

```python
# коэффициент корреляции чуть больше, чем в методе Пирсона, однако корреляция всё ещё слабая прямая
# p-value значительно уменьшился, и теперь мы можем доверять данному коэффициенту, т.к. p-value чуть больше 0.05
```
